### Problem Statement:-

1. Generate a random integer from a to b. 
2. You and your friend have to guess a number between two numbers, a and b. 
3. a and b should be inputs taken from the user only. 

Your friend is player 1 and plays first. He will have to keep choosing the number, and your program must tell whether the number is greater than the actual number or less than the actual number. Log the number of trials it took your friend to arrive at the number. 

You play the same game, and then the person with the minimum number of trials wins! 
Randomly generate a number after taking a and b as input and donâ€™t show that to the user.

#### Input :
Enter start (a) : 4

Enter end (b) : 13

#### Output :
Player1 :

Please guess the number between 4 and 13

5

Wrong guess a greater number again

8

Wrong guess a smaller number again

6

Correct, you took 3 trials to guess the number

Player 2:

Correct, you took 7 trials to guess the number

Player 1 wins!

### My initial take

```python
a = int(input("Enter your start : "))
b = int(input("Enter your end : "))
actual_number = int(input(f"Imagine any number between {a} and {b} : "))
trials = 1

for i in range(a,b):
    guess = int(input(f"Please guess any number between {a} and {b} : "))
    while guess != actual_number:
        if guess < actual_number:
            print("Ah ! Wrong guess. The number you've guessed is smaller. please try again.")
            trials += 1
            break
        elif guess > actual_number:
            print("Ah ! Wrong guess. The number you've guessed is greater. please try again.")
            trials +=1
            break
    if guess == actual_number:
        print(f"Correct! You took {trials} trials to guess the right number.")
        break
```

### Solution on Code with Harry

```python
import random

# defining the rules of our game
def guess_game(a, b, actual):
    guess = int(input(f"Guess a number between {a} and {b} : \n"))
    trials = 1
    
    while guess != actual:
        if guess < actual:
            guess = int(input("Wrong! Enter a bigger number :- \n"))
            trials +=1
        else:
            guess = int(input("Wrong! Enter a smaller number :- \n"))
            trials +=1
            
    print(f"You've guessed the right number after {trials} gueses.", end = "\n\n")
    return trials

# this will declare the range 
a = int(input("Enter start :- \n"))
b = int(input("Enter end :- \n"))

# to randomly choose any number between the given range
actual_1 = random.randint(a, b)
actual_2 = random.randint(a, b)

# game starts here
print("Player 1 turn :- ")
g1 = guess_game(a,b,actual_1)
print("Player 2 turn :- ")
g2 = guess_game(a,b,actual_2)

# final results
if g1 < g2:
    print("Player 1 won !")
elif g2 < g1:
    print("Player 2 won !")
else:
    ("Its a Draw !")
    
print(f"The number for 1st Player was {actual_1} and for 2nd Player was {actual_2}")
```

### Making it more dynamic

```python
import random

print("Welcome to 'Guess the Number!' \n")

# defining the rules
def guess_game(a, b, actual):
    
    trial_no = 1
    chances = 6
    
    while chances >= 6:
        guess = int(input(f"Guess a number between {a} and {b} : \n{chances} chances left! \n"))
        while guess != actual:
            if chances == 1:
                print("No chances left! Game Over !")
                break
                
            if chances == 2:
                chances -=1
                guess = int(input(f"Wrong! Enter a bigger number :- \nCareful, last chance!\n"))
                trial_no +=1
                if guess != actual:
                    if chances == 1:
                        print("No chances left. Game Over. \n")
                        break
                else:
                    print(f"You've guessed the right number after {trial_no} gueses.", end = "\n")
                    
            if guess < actual:
                chances -=1
                guess = int(input(f"Wrong! Enter a bigger number :- \n{chances} chances left! \n"))
                trial_no +=1
                      
            elif guess > actual:
                chances -=1
                guess = int(input(f"Wrong! Enter a smaller number :- \n{chances} chances left! \n"))
                trial_no +=1
        
        print(f"\nYou took {trial_no} trial/s \n") 
        
    return trial_no

# this will declare the range 
a = int(input("Enter start :- \n"))
b = int(input("Enter end :- \n"))

# to randomly choose any number between the given range
actual_1 = random.randint(a, b)
actual_2 = random.randint(a, b)

# game starts here
print("\nPlayer 1 turn :- ")
g1 = guess_game(a,b,actual_1)
print("Player 2 turn :- ")
g2 = guess_game(a,b,actual_2)

# final results
if g1 < g2:
    print("Player 1 won!")
elif g2 < g1:
    print("Player 2 won!")
else:
    ("Its a Draw !")
    
print(f"The number for 1st Player was {actual_1} and for 2nd Player was {actual_2}.")
```
